```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# our performance metrics results for classification

recall = np.array([0.01,0.3,0.4,0.5,0.7,0.8])     #recall rates from different model tests
precision = np.array([0.9,1.4,1.5,1.7,1.8,1.9])  #precision rates from different model tests
```

```python
RecallTimesPrecision = recall * precision
RecallTimesPrecision
```

```python
RecallPlusPrecision = recall + precision
RecallPlusPrecision
```

```python
# we use harmonic mean to punish extreme values in our results
# to make outliars more visible

harmonic_mean = 2*(RecallTimesPrecision/RecallPlusPrecision)
harmonic_mean
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(recall, label='recall')
ax.plot(precision, label='precision')
ax.plot(harmonic_mean, label='harmonic_mean')
ax.legend()

# observe that the extreme first value from 'recall' array is punished 
# (that is the objective of using harmonic mean)
```
